## Discretization

```python
def dummy_one_hot(X,key):
    dummies=pd.get_dummies(X[key],prefix=key)
    df=X.join(dummies)
    df.drop([key],axis=1,inplace=True)
    return df


def discret(X):
    X['age']=pd.cut(X['age'],10,include_lowest=True)
    X['fnlwgt']=pd.cut(X['fnlwgt'],10,include_lowest=True)    
    X['capital_gain']=pd.cut(X['capital_gain'],np.arange(0,100100,1000),include_lowest=True)    
    X['capital_loss']=pd.cut(X['capital_loss'],np.arange(0,4500,100),include_lowest=True)
    X['hours_per_week']=pd.cut(X['hours_per_week'],10,include_lowest=True)
    X=dummy_one_hot(X,'age')
    X=dummy_one_hot(X,'fnlwgt')    
    X=dummy_one_hot(X,'capital_gain')
    X=dummy_one_hot(X,'capital_loss')
    X=dummy_one_hot(X,'hours_per_week')


    
    return X


df_X_train= discret(df_X_train)
df_X_train.head(10)
```

```python
X_train = np.array(df_X_train)
X_train = X_train.reshape(-1,X_train.shape[1])

y_train = np.array(df_y_train)
y_train = y_train.reshape(-1,1)
```

# Implement Generative Model method:

## data preprocessing : normalization

```python
import pandas as pd
import numpy as np
from numpy.linalg import inv

# class generative_model()

DIR_MAC="/Users/haipengwu/Codings/Machine Learning/HW2/"
DIR_1080="C:\\Users\\WIN10\\Codings\\Machine Learning\\HW2\\"

FILE_X_train="X_train"
FILE_y_train="Y_train"
FILE_X_test="X_test"

#put directory info inside Run_machine
RUN_MACHINE =DIR_1080


#106 features for X_train
df_X_train = pd.read_csv(RUN_MACHINE+FILE_X_train, encoding='unicode_escape')
# y =0 means income <=50k, y=1 means income >50k
df_y_train  = pd.read_csv(RUN_MACHINE+FILE_y_train, encoding='unicode_escape')

df_X_test = pd.read_csv(RUN_MACHINE+FILE_X_test, encoding='unicode_escape')



X_train = np.array(df_X_train)
X_train = X_train.reshape(-1,X_train.shape[1])

y_train = np.array(df_y_train)
y_train = y_train.reshape(-1,1)

from sklearn.preprocessing import scale,normalize,minmax_scale


def norm(X):

#     X = normalize(X, axis=0,norm='l2')
    X = minmax_scale(X)
#     X = scale(X)

    return X

X_train = norm(X_train)


X_test = np.array(df_X_test)
X_test = X_test.reshape(-1, X_test.shape[1])
X_test = norm(X_test)

X_test.shape
```

```python
print(X_train.shape,y_train.shape)
class_0_index=[]
class_1_index=[]

#class_0 and class_1's index
for i in range(y_train.shape[0]):
    if y_train[i][0]==0:
        class_0_index.append(i)
    else:
        class_1_index.append(i)
print(len(class_0_index),len(class_1_index))



#X_train class 0 and class 1 data

X_train_class_0 = X_train[class_0_index]
X_train_class_1 = X_train[class_1_index]

mean_class_0 = np.mean(X_train[class_0_index],axis = 0)
mean_class_1 = np.mean(X_train[class_1_index],axis = 0)

n = X_train.shape[1]
cov_class_0 = np.zeros((n,n))
cov_class_1 = np.zeros((n,n))

for i in range(X_train_class_0.shape[0]):
    
    cov_class_0 += np.dot(np.transpose([X_train_class_0[i] - mean_class_0]), [(X_train_class_0[i] -mean_class_0)]) / X_train_class_0.shape[0]

for i in range(X_train_class_1.shape[0]):
    cov_class_1 += np.dot(np.transpose([X_train_class_1[i] - mean_class_1]), [(X_train_class_1[i] - mean_class_1)]) / X_train_class_1.shape[0]
            
cov = (cov_class_0*X_train_class_0.shape[0] + cov_class_1*X_train_class_1.shape[0]) / (X_train_class_0.shape[0] + X_train_class_1.shape[0])

print(cov)

```

## mean,cov to generate 'W, b'

```python
delta_mean = mean_class_0 - mean_class_1
# delta_mean =delta_mean.reshape(X_train.shape[1],1)
print(delta_mean.shape,cov.shape)
W = np.matmul(np.transpose(delta_mean),inv(cov))
b = (- 0.5)* (mean_class_0).dot(inv(cov)).dot(mean_class_0)\
            + 0.5 * (mean_class_1).dot(inv(cov)).dot(mean_class_1)\
            + np.log(float(mean_class_0.shape[0]) / mean_class_1.shape[0]) 
print(W.shape,b.shape)
```

## call sigmoid(W * X_test + b)

```python
def sigmoid(z):
    return np.clip(1/(1+np.exp(-z)),1e-8,1-1e-8)
def f_prob(x):
    arr = np.empty([x.shape[0],1],dtype=float)
    for i in range(x.shape[0]):
        arr[i][0] = sigmoid(np.dot(x[i,:],W) + b)
    return arr

results =np.empty([X_test.shape[0],1],dtype=float)
results = f_prob(X_test)
for result in results:
    if result[0]> 0.5:
        result[0] = 0
    else:
        result[0] = 1
    
```

```python
y_test_pred =np.round(results)
print(y_test_pred)
```

```python
id_index =[int(i) for i in range(1,16282)]
id_index=np.array(id_index)
id_index=id_index.reshape(-1,1)
results = np.stack([id_index, y_test_pred],axis=1)
results =results.reshape(-1,2,)
results =results.astype(np.int)

import time

submission_file_name = RUN_MACHINE + "sample_submission_wu{}.csv".format(time.time())

results = pd.DataFrame(results,columns=['id','label'])

results.to_csv(submission_file_name,index=False)
pd.read_csv(submission_file_name)
```
